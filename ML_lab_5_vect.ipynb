```python
import numpy as np
import pandas as pd
import os
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(add_indicator=False,copy=True, missing_values=np.nan,strategy='constant',fill_value='',verbose=0)
imputer
```

```python
vectorizer = CountVectorizer(stop_words='english')
vectorizer
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/movies_metadata1.csv')
data.head()
```

```python
title = data['original_title']
title.head()
```

```python
title.isnull().any()
```

```python
title1 = imputer.fit_transform(np.array(title.values).reshape(-1,1))
title1
```

```python
title1.shape
```

```python
pd.isnull(title1).any()
```

```python
title1 = title1[:,0]
title1.shape
```

```python
vectorizer.fit(title1)
```

```python
vector = vectorizer.transform(title1)
vector
```

```python
len(vectorizer.vocabulary_.keys())
```
